```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
# Strip whitespace and convert to int
df['Survived'] = df['Survived'].astype(str).str.strip().astype(int)
df['Pclass'] = df['Pclass'].astype(str).str.strip().astype(int)
```

```python
sns.countplot(x='Survived', data=df)
plt.show()
# In THis Plot 1 is Showing how many people are Survived and 0 is how many are dead
# This Plot show us how many people are survived by count its vlaues how many time occur and show how many are dead
```

```python
# Lets Apply CountPlot on Pclass And get how many time each catagory occur in our data set
sns.countplot(x='Pclass',data=df)
plt.show()
```

```python
# Checking The Gender Compresion how much are male or how much female
sns.countplot(x='Sex',data=df)
# From This plot we can clearly see that the male is more than female in this ship
```

### Pie Chart

```python
# Here we are now using pie chart to see the data in percentage or in area of circle form
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
# Here we are using value count its will count the value of the how much numbers of males and females and then plot on pie chart to show the 
# bold area according to value_counts()
```

### Histogram

```python
# We can use histogram to deal with numerical data to get the frequency at how much is values are at that point 
# E.G we age coloumn tell how much people are in between 20 to 30 age like that
plt.hist(df['Age'],bins=5)
```

### DisPlot

```python
# DisPlot is same as the Histogram plot but its geneate line on histogram and probility on y-axis to that show how much 
# propability is like age like that
sns.distplot(df['Age'])
# As we can see that on y-axis this displot show the probabilty of the the age 
```

### BOXPLOT

```python
sns.boxplot(df['Age'])
# Box Plot Show us highlighting the median, quartiles, and outliers. Here we can see that the mid is medium value the tow lines are drawan
# And At the upper that are outlier 
```
